```python
message="""\nProgram Options:
1) Add book
2) print library books 
3) print books by prefix
4) Add user
5) Borrow book 
6) Return book
7) print users that borrowed book
8) print users
9) End Program
"""
```

```python
class Book():
    def __init__(self,id,name,quantity):
        self.id=id
        self.name=name
        self.quantity=quantity
        self.borrowed=0
    def to_string(self):
        print(f'Book name: {self.name}, id: {self.id},  current quantity: {self.quantity}, currently borrowed: {self.borrowed}')
        
```

```python
class Library():
    
    def __init__(self):
        self.books=[]
        self.users=[]
        self.users_borrowed_books={}
        
    def add_book(self,id,name,quantity):
        book=Book(id,name,quantity)
        self.books.append(book) 
        
    def add_user(self,id,name):
        user=User(id,name)
        self.users.append(user) 
        
    def print_search_pre(self,prefix):
        flag=0
        for book in self.books:
            if book.name.lower().startswith(prefix.lower()):
                book.to_string()
                flag=1
        if not flag:
            print('No book found')
            
    def print_all_books(self):
        for book in self.books:
            book.to_string()
    
            
    def print_all_users(self):
        for user in self.users:
            user.to_string()
            try:
                print('Borrowed books:')
                self.users_borrowed_books.get(user).to_string()
            except:
                print('',end='')
                
    def print_users_borrowed_book(self,bookname):
        print('list of users borrowed this book:')
        for k,v in self.users_borrowed_books.items():
            if v.name==bookname:
                k.to_string()
                
        
                
    def borrow_book(self,username,bookname):
        for user in self.users:
            if user.name==username:
                for book in self.books:
                    if book.name==bookname:
                        if book.quantity > 0:
                            book.quantity -= 1
                            book.borrowed += 1
                            self.users_borrowed_books[user]=book
                        else:
                            print('No more copies left of this book to borrow.')
                        return     
                print('book name not found')
        print('user not found')
        
    def return_book(self,bookname,username):
        for key,value in self.users_borrowed_books.items():
            if key.name==username and value.name==bookname:
                value.quantity += 1
                value.borrowed -= 1
                self.users_borrowed_books.pop(key)
                return
                
        print("This user didn't borrow this book")
                        
                        
                    
    
            
        
```

```python
class User():
    def __init__(self,id,name):
        self.id=id
        self.name=name
        
    def to_string(self):
        print(f'user name: {self.name},id: {self.id} ')
```

# Main Program

```python
lib=Library()
while True:
    print(message)
    case=int(input('\nEnter your choice: '))
    
    if case == 1:
        id=int(input('Enter book id: '))
        name=input('Enter book name: ')
        quantity=int(input('Enter book quantity: '))
        lib.add_book(id,name,quantity)
        
    elif case == 2:
        lib.print_all_books()
        
    elif case == 3:
        prefix=input('Enter prefix: ')
        lib.print_search_pre(prefix)
        
    elif case == 4:
        id=int(input('Enter user id: '))
        name=input('Enter user name: ')
        lib.add_user(id,name)
        
    elif case == 5:
        username=input('Enter username: ')
        bookname=input('Enter book name: ')
        lib.borrow_book(username,bookname)
        
    elif case == 6:
        username=input('Enter username: ')
        bookname=input('Enter book name: ')
        lib.return_book(bookname,username)
        
    elif case == 7:
        bookname=input('Enter book name: ')
        lib.print_users_borrowed_book(bookname)
        
    elif case == 8:
        lib.print_all_users()
        
    elif case == 9:
        break
    
    else:
        print('Invalid Input')
```
